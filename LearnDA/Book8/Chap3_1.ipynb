```python
## 习题1
a_1=100
b_1=55
c_1=299
d_1=399
e_1=86
def printAns(arg):
    print(arg)
    print(bin(arg))
    print(oct(arg))
    print(hex(arg))
    print("\n")
printAns(a_1)
printAns(b_1)
printAns(c_1)
printAns(d_1)
printAns(e_1)
```
