## 访问数据集中的某一列可以用两种方法
Dataframe中的某一列均为Series

1.
```python
dataset["column_name"]
```
2. 链式方法
不建议采用此种方法，原因如下：
其一，如果对应的列名包含特殊字符，这种方法可能无效，如果列名为CountryName，那么无法通过代码retail_data.Country Name来访问对应列；
其二，如果列名与DataFrame的方法重名，那么也会无效，如果数据中某列名为mean，此时也是无法采用该方法访问列的
```python
dataset.column_name
```
