# 函数与代码复用

## 函数的定义

在前面已经接触过python中函数的定义，python使用def关键字来对函数进行定义

### 实例5.1

````python
# HappyBirthday.py
def happy():
    print("Happy Birthday to You!")
def happyB(name):
    happy()
    happy()
    print("Happy Birthday,dear {}!".format(name))
    happy()
happyB("Mike")
print()
happyB("Lily")
````

## lambda函数

python的33个保留字当中的一个是lambda，它用于定义匿名函数

````python
f=lambda x,y:x+y
type(f)
````

````python
f(10,12)
````

一般的匿名函数的定义方式如下
```python
<函数名>=lambda <参数列表> : <表达式>
```

## 函数的参数传递

### 可选参数和可变数量参数


````python
def dup(str,times=2):
    print(str*times)
dup("knock~")
````

````python
dup("knock~",4)
````

上面给出的示例当中，定义的dup函数的times参数给定了默认值，在函数调用时，如果其实际参数缺省，则将使用默认值代替。由于调用时，需要按照顺序输入参数，可选参数必须定义在非可选参数的后面。
另外函数定义时还可以添加可变数量的参数，通过在参数前面添加星号，可变数量参数的位置也必须出现在参数列表的最后。函数在调用时，这些参数会被当做元组类型传入函数。

````python
def vfunc(a,*b):
    print(type(b))
    for n in b:
        a+=n
    return a
vfunc(1,2,3,4,5)
````

### 参数的位置与名称传递
为了解决仅看参数难以理解参数含义的问题，python提供了按照形参名称输入实参的方式，语法如下：
```python
result=func(x2=4,y2=5,z2=6,x1=1,y1=2,z1=3)
```

# 函数的定义

````python
def fact():
    print("也是一个函数")
fact()
````

````python
def fact(n):
    s=1
    for i in range(1,n+1):
        s*=i
    return s
print(fact(10))
````

````python
def fact(n,m=1):
    s=1
    for i in range(1,n+1):
        s*=i
    return s//m
fact(4)
````

````python
def fact(n,*b):
    s=1
    for i in range(1,n+1):
        s*=i
    for item in b:
        s*=item
    return s
fact(5,1,2,3)
````

n被设置为5，这是fact()函数的第一个参数，b被设置为 (1, 2, 3)，这是fact()函数的可变位置参数，$fact(5,1,2,3)$的结果将是$5!*1*2*3$的乘积

````python
n,s=10,100
def fact(n):
    global s
    for i in range(1,n+1):
        s*=i
    return s
print(fact(n),s)
````
