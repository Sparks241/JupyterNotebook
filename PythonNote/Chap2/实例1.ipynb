## 温度转换v1.0
使用条件语句，仅能判断一次

````python
#TempConvert.py
#v1.1
TempStr=input("请输入带有温度的符号值：")#从键盘获取用户输入的待转换的温度
if TempStr[-1] in ['F','f']:#使用if语句判断输入的温度的类型为华氏还是摄氏
    C=(eval(TempStr[0:-1])-32)/1.8
    print("转换后的温度是{:.2f}C".format(C))
elif TempStr[-1] in ['C','c']:#分别按照公式进行转换，并且使用格式控制输出
    F=1.8*eval(TempStr[0:-1])+32
    print("转换后的温度是{:.2f}F".format(F))
else:
    print("输入格式错误")#使用else对错误进行处理
````

## 温度转换v1.1
保持原功能主题上不变的前提下，增加了循环以实现对于温度的多次转换

````python
#TempConvert.py
#v1.1
print("--------温度转换开始--------")
TempStr=input("请输入带有温度的符号值：")#从键盘获取用户输入的待转换的温度（在循环开始之前需要进行额外说明）
while TempStr[-1] not in ['N','n']:
    if TempStr[-1] in ['F','f']:#使用if语句判断输入的温度的类型为华氏还是摄氏
        C=(eval(TempStr[0:-1])-32)/1.8
        print("转换后的温度是{:.2f}C".format(C))
    elif TempStr[-1] in ['C','c']:#分别按照公式进行转换，并且使用格式控制输出
        F=1.8*eval(TempStr[0:-1])+32
        print("转换后的温度是{:.2f}F".format(F))
    else:
        print("输入格式错误")#使用else对错误进行处理
    TempStr=input("请输入带有符号的温度值：")
print("--------温度转换结束--------")
````

1. not关键字

not关键字表示对结果取反
