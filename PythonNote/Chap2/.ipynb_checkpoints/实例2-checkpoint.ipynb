## 蟒蛇绘制

python中的turtle库可以用于绘制一些图形，这里给出一段使用turtle库函数绘制蟒蛇的示例

````python
#PythonDraw.py
#v1.0
import turtle
from PIL import Image
screen = turtle.Screen()
screen.setup(width=600, height=600)
turtle.setup(650,350,200,200)
turtle.penup()
turtle.fd(-250)
turtle.pendown()
turtle.pensize(40)
turtle.pencolor("purple")
turtle.seth(-40)
for i in range(4):
    turtle.circle(40,80)
    turtle.circle(-40,80)
turtle.circle(40,80/2)
turtle.fd(40)
turtle.circle(16,180)
turtle.fd(40*2/3)
canvas = screen.getcanvas()
# 保存到文件
canvas.postscript(file="python01.eps")
# 使用Pillow将EPS转换为PNG
img = Image.open("python01.eps")
img.save("python01.png")
turtle.done()
````

### 知识点

1. 引用函数库

    - 引用函数库可以采用如下的形式
    ```python
    import <库名>
    import turtle
    <库名>.<函数名>(<函数参数>)
    ```
    - 亦可以采用如下的方式
    ```python
    from <库名> import <函数名，函数名，函数名>
    from <库名> import *
    ```
使用第二种库引用方式修改实例二

````python
#PythonDraw.py
#v1.1
from turtle import *
from PIL import Image
screen = turtle.Screen()
screen.setup(width=600, height=600)
turtle.setup(650,350,200,200)
turtle.penup()
turtle.fd(-250)
turtle.pendown()
turtle.pensize(40)
turtle.pencolor("purple")
turtle.seth(-40)
for i in range(4):
    turtle.circle(40,80)
    turtle.circle(-40,80)
turtle.circle(40,80/2)
turtle.fd(40)
turtle.circle(16,180)
turtle.fd(40*2/3)
canvas = screen.getcanvas()
# 保存到文件
canvas.postscript(file="python01.eps")
# 使用Pillow将EPS转换为PNG
img = Image.open("python01.eps")
img.save("python01.png")
turtle.done()
````

### 知识点
2. turtle库元素语法分析
    - 绘图坐标体系：开始绘制时，小海龟将位于画布正中央
    - 定义图坐标体系
    ```python
    turtle.setup(width,height,startx,starty)
    t.setup(650,350,200,200)
    ```
    - 参数说明
        - width与height参数为整数则表示像素值，各参数若为小数，表示窗口参数与屏幕参数比例
        - startx为窗口左侧与屏幕左侧距离，值为None则位于中央
        - starty同理，为上侧的间距
3. 基本操作
    - penup（up或者pu）与pendown（pd或者down），分别控制海龟悬空和落下
    - pensize（width）：控制绘制的笔触粗细
    - pencolor：控制笔触的颜色
        - colorstring："purple","red","blue"
        - (r,g,b)：颜色对应的RGB数值
    - fd（forward）函数：前进
        - distance：前进的像素值
    - seth（setheading）：设置前进的绝对角度，以正东为0度
    - circle
    ```python
    turtle.circle(radius,extext=None)
    ```
    - 参数说明
        - radius为半径，值为正数绘制在左侧，负数绘制在右侧
        - extent为角度

## 蟒蛇绘制v1.2

````python
#PythonDraw.py
#v1.2
import turtle
from PIL import Image
screen = turtle.Screen()
screen.setup(width=600, height=600)
def drawSnake(radius,angle,length):
    turtle.seth(-40)
    for i in range(length):
        turtle.circle(radius,angle)
        turtle.circle(-radius,angle)
    turtle.circle(radius,angle/2)
    turtle.fd(40)
    turtle.circle(16,180)
    turtle.fd(40*2/3)
turtle.setup(650,350,200,200)
turtle.penup()
turtle.fd(-250)
turtle.pendown()
turtle.pensize(40)
turtle.pencolor("purple")
drawSnake(40,80,4)

canvas = screen.getcanvas()
# 保存到文件
canvas.postscript(file="python02.eps")
# 使用Pillow将EPS转换为PNG
img = Image.open("python02.eps")
img.save("python02.png")
turtle.done()
````

````python
# PythonDraw.py
# v1.3
from turtle import *
from PIL import Image

screen = Screen()
setup(650, 350, 200, 200)
penup()
fd(-250)
pendown()
pensize(25)
colormode(255)  # 将颜色模式设置为RGB范围0-255
pencolor(255, 192, 103)  # 设置画笔颜色为RGB(255, 192, 103)
seth(-40)
for i in range(4):
    circle(40, 80)
    circle(-40, 80)
fd(40)
circle(16, 180)
fd(40 * 2/3)
canvas = screen.getcanvas()
canvas.postscript(file="python03.eps")
# 使用Pillow将EPS转换为PNG
img = Image.open("python03.eps")
img.save("python03.png")
done()
````
