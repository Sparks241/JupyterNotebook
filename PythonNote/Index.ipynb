# Python笔记

## 说明

### 完成方式

本次Python笔记主体使用Jupyter Notebook来完成，文字性的部分使用Markdown语法完成，代码部分使用jupyter内置的python内核来完成。

### 图像处理

使用pillow库以及ghostpillow将python绘制的图像保存至本地，使用Markdown语法将图片插入jupyterNotebook。

### 文件格式转换

用pandoc将ipynb文件转换为markdown文件，接着将md文件转换为docx文件

## 文件组织方式

由于存在文件转换过程，每个文件夹下均有与ipynb文件同名的md文件和docx文件，后两者在此省略，仅列出ipynb文件

```markdown
PythonNote
    |----Chap1
        |---Codes.ipynb
        |---turtle001.eps
        |---turtle001.png
    |----Chap2
        |---实例1.ipynb
        |---实例2.ipynb
    |----Chap3
    |----
    |----
    |----Chap6
        |---CombinedDataType
            |---CombinedDataType.ipynb
    |----Others
    |----Index.ipynb
```
