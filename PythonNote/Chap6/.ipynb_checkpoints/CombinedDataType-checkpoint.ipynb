# jieba库的使用

jieba是一个优秀的中文分词工具，用于对中文文本进行分词处理
```python
import jieba
s="中国是一个伟大的国家"#导入jieba库，用于中文分词。
jieba.lcut("中国是一个伟大的国家")
jieba.lcut("中国是一个伟大的国家",cut_all=True)#使用jieba的lcut方法对字符串进行精确模式的分词，返回一个分词后的列表。
jieba.lcut_for_search("中华人民共和国是伟大的")
jieba.add_word("小小",cut_all=True)
s2="我是一只小小鸟"
jieba.lcut(s2,cut_all=True)
```

````python
import os
def getText():
    txt=open("ham.txt","r").read()#使用open()函数打开名为"ham.txt"的文件，并使用read()方法读取其中的内容，并将内容赋给变量txt
    txt=txt.lower()#使用lower()方法将文本内容转换为小写
    for ch in '!"#$%&()*+,-./:;<=>?@[\\]^^_‘{|}~':#使用for循环遍历特殊字符，然后使用replace()方法将这些特殊字符替换为空格
        txt=txt.replace(ch," ")
    return txt
hamletTxt=getText()#调用getText()函数，将处理后的文本内容保存在变量hamletTxt中
words=hamletTxt.split()#使用split()方法将文本内容按照空格分割成单词，并将结果保存在列表words中
counts={}
for word in words:#遍历words列表中的每个单词，统计每个单词出现的次数，并将结果保存在字典counts中
    counts[word]=counts.get(word,0)+1
items=list(counts.items())
items.sort(key=lambda x:x[1],reverse=True)#将counts字典中的键值对转换为列表，并按照单词出现次数进行排序，从高到低排序
for i in range(10):
    word,count=items[i]#遍历排序后的前十个元素，分别输出单词和出现次数
    print("{0:<10}{1:>5}".format(word,count))
````

````python
txt="I am am a a boOK, DKFJ! &IS"
txt=txt.lower()#将字符串txt中的所有字母转换为小写
for ch in ",!&":#使用for循环遍历逗号、叹号和和符号，并使用replace方法将这些字符替换为空格
    txt=txt.replace(ch," ")
print(txt)
lt=txt.split()#使用split方法将处理后的字符串按照空格分割成单词，并将结果保存在列表lt中
counts={}
for word in lt:
    counts[word]=counts.get(word,0)+1#遍历列表lt中的每个单词，统计每个单词出现的次数，并将结果保存在字典counts中
print(counts)
````

## 平均值与方差的计算方法

````python
ls=[12,23,33,43,21,45,66]#定义一个存放n个数字
s=0
for i in ls:
    s=s+i
ave=s/len(ls)
print(ave)#输出平均值
````

````python
s=0
for i in ls:
    s=s+(i-ave)*(i-ave)
s=s/len(ls)
s=pow(s,1/2)
print(s)#输出方差
````

# 示例9：基本统计值计算


````python
#CalStatistics.py
#v1.0
from math import sqrt#导入了数学模块中的平方根函数 sqrt
def getNum():#定义获取数字的函数
    nums=[]#创建了一个空列表 nums，用于存储用户输入的数字
    iNumStr=input("请输入数字（直接输入回车退出）")#使用 input() 函数获取用户输入的字符串
    while iNumStr != "":#用户输入空字符串时，循环结束，函数返回 nums 列表
        nums.append(eval(iNumStr))#对nums进行附加数字
        iNumStr=input("请输入数字（直接回车退出）")
    return nums
def mean(numbers):
    s=0.0
    for num in numbers:#使用 for 循环遍历列表中的每个数字，将其累加到 s
        s=s+num
    return s/len(numbers)
def dev(numbers,mean):#定义了一个名为 dev() 的函数，用于计算列表中数字的标准差
    sdev=0.0
    for num in numbers:#使用 for 循环遍历列表中的每个数字，计算其与平均值的差的平方，并将结果累加到 sdev 中
        sdev=sdev+(num-mean)**2
    return sqrt(sdev/(len(numbers)-1))
def median(numbers):#定义了一个名为 median() 的函数，用于计算列表中数字的中位数
    new=sorted(numbers)
    size=len(numbers)
    if size % 2 == 0:#判断列表长度是否为偶数，如果是偶数，则取中间两个数的平均值作为中位数；如果是奇数，则直接取中间的数作为中位数
        med=(new[size//2-1]+new[size//2])/2
    else:
        med=new[size//2]
    return med
n=getNum()
m=mean(n)
print("平均值：{},标准差：{:.2},中位数：{}.".format(m,dev(n,m),median(n)))
````

````python
# 文件的使用
## 实例DiffTextBin.py
tf=open("f.txt","rt",encoding="UTF-8")#open() 函数打开文件"f.txt"，以文本模式（"rt"）读取，并指定编码为UTF-8
print(tf.readline())#readline() 方法读取文件的第一行，并将其打印输出
tf.close()
bf=open("f.txt","rb")#open() 函数以二进制模式（"rb"）打开同一个文件"f.txt"，文件被解析为字节流
print(bf.readline())#readline() 方法读取文件的第一行，并将其打印输出
bf.close()
````
