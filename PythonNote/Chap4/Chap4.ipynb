# 循环结构

```python
for s in "BIT":
    print("循环进行中："+s)
else:
    s="循环正常结束"
print(s)
```

使用一个while循环来遍历字符串s中的每个字符，循环条件是idx的值小于字符串s的长度，循环体中，输出当前遍历到的字符，通过索引idx访问字符串中的字符，每次循环结束，idx自增1

```python
s,idx="BIT",0
while idx<len(s):
    print("循环进行中："+s[idx])
    idx+=1
else:
    s="循环正常结束"
print(s)
```

定义了一个内层循环，使用range(10)创建了一个包含10个整数的迭代器。内层循环会执行10次，检查当前字符s是否等于"I"，如果是，则使用break语句跳出内层循环

```python
for s in "BIT":
    for i in range(10):
        print(s,end="")
        if s=="I":
            break
```

```python
for s in "PYTHON":
    if s=="T":
        continue
    print(s,end="")
```

```python
for s in "PYTHON":
    if s=="T":
        break
    print(s,end="")
```

continue仅结束本次的循环，但不跳出当前循环，而break会跳出当前层次的循环。

```python
for s in "PYTHON":
    if s=="T":
        continue
    print(s,end="")
else:
    print("正常退出")
```

```python
for s in "PYTHON":
    if s=="T":
        break
    print(s,end="")
else:
    print("正常退出")
```

上面演示的是for与while循环中的else的扩展用法。执行else语句块的条件是循环正常遍历了所有内容，或者条件不成立而结束。

```python
try:  # 异常处理 当用户没有输入整数的时候
    num=eval(input("请输入一个整数："))
    print(num**2)
except:
    print("输入不是整数")
```

```python
try:
    alp="ABCDEFGHIJKLMNOPQRSTUVWXYZ"
    idx=eval(input("请输入一个整数："))
    print(alp[idx])
except NmaeError:
    print("输入错误，请输入一个整数！")
else:
    print("没有发生异常")
finally:
    print("程序执行完毕，不知道是否发生了异常")
```

下面给出的这段代码使用了try-except块，通过调用集合A的pop()，弹出并输出集合A中的每个元素，直到集合A为空。异常捕获块（except）用于捕获当集合为空时的异常，并使用pass语句来忽略异常，保证程序正常执行。

```python
A={"p","y",123}
B=set("pypy123")
print(A)
print(B)
print(A-B)#分别打印了集合的差集、交集和对称差集
print(A&B)
print(A^B)
print(B-A)
for item in A:#使用for循环遍历集合A中的每个元素
    print(item)
try:
    while True:
        print(A.pop(),end="")
except:
    pass
```

# 实例pi的计算

```python
from random import random
from math import sqrt
from time import perf_counter
DARTS=1000
hits=0.0
perf_counter()#perf_counter() 函数用于记录程序开始的时间
for i in range(1,DARTS+1):
    x,y=random(),random()
    dist=sqrt(x**2+y**2)
    if dist<=1.0:
        hits=hits+1
pi=4*(hits/DARTS)
print("Pi值是{}".format(pi))
print("运行时间是：{:.5f}s".format(perf_counter()))
```

代码中的循环执行了DARTS次迭代，每次迭代生成一个随机点$(x, y)$，并计算该点到原点的距离dist。如果距离小于等于1.0，表示该点落在了以原点为圆心、半径为1的圆内，因此将 hits 的计数加1。

根据蒙特卡罗方法的原理，利用落在圆内的次数 hits 与总投掷次数 DARTS 的比例来估算圆周率π的值

# 实例天天向上

```python
#DayDayUp.py
#v1.0
def dayUP(df):
    dayup=1
    for i in range(365):
        if i%7 in[6,0]:
            dayup=dayup*(1-0.01)
        else:
            dayup=dayup*(1+df)
    return dayup
dayfactor=0.01
while dayUP(dayfactor)<37.78:
    dayfactor+=0.01
print("{:.3f}".format(dayfactor))
```

函数 dayUP(df) 接受一个参数 df，表示每日努力指数的增长率。在函数中，通过循环模拟了一年中每天的努力情况,如果是周末（即 $i%7$的余数是6或0），则努力指数按照每天的$1\%$的比例下降（即一周中每周末休息两天），否则，努力指数按照每日的 df 的比例增长。
