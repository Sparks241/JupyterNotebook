# 几个简单的Python小程序

## 微实例1

```python
radius=25#定义圆的半径为25
area=3.1415*radius*radius#定义圆的面积使用公式计算
print(area)
print("{:.2f}".format(area))#使用格式控制输出，仅输出两位小数
```

## 微实例2

```python
name=input("输入姓名:")
print("{}同学，学好Python，前途无量！".format(name))#格式控制的进阶用法，format括号中的量将会输出，引号内类似于C语言
print("{}大侠，学好Python，大展拳脚！".format(name[0]))#输出第一个字符，属于正向递增，最左侧的字符从0开始
print("{}哥哥，学好Python，前途无量！".format(name[1:]))#输出name字符串第一个字符以后的内容
```

## 微实例3

```python
a,b=0,1
while a<1000:#循环：当a小于1000时执行下列语句
    print(a,end=',')#输出a，并且以逗号作为分隔
    a,b=b,a+b#将b的值赋值给a，将a与b的和赋值给b
    #本段函数最终实现输出1000以内的裴波那契数列，0和1为裴波那契数列最开始的两个数
```

## 微实例4：同切圆的绘制

```python
import turtle as t
from PIL import Image
# 设置画布
screen = t.Screen()
screen.setup(width=600, height=600)
t.pensize(2)
t.circle(10)
t.circle(40)
t.circle(80)
t.circle(160)
canvas = screen.getcanvas()
# 保存到文件
canvas.postscript(file="turtle001.eps")
# 使用Pillow将EPS转换为PNG
img = Image.open("turtle001.eps")
img.save("turtle001.png")
# 结束turtle绘图
t.bye()
```

```python
from turtle import *
from PIL import Image
screen=Screen()
screen.setup(width=600, height=600)
color('red','yellow')
begin_fill()
while True:
    forward(200)
    left(170)
    if abs(pos())<1:
        break
end_fill()
canvas = screen.getcanvas()
# 保存到文件
canvas.postscript(file="flower1.eps")
# 使用Pillow将EPS转换为PNG
img = Image.open("flower1.eps")
img.save("flower1.png")
# 结束turtle绘图
done()
```
