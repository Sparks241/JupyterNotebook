# 数字类型

## 概述

python语言提供的数字类型有：整数、浮点数、复数

## 整数

整数类型理论上是$[-\infty,\infty]$，实际上受限于运行程序的计算机的内存大小。可以使用pow(x,y)函数（用于计算$x^y$）来测试整数类型的取值范围

```python
pow(2,100)
```

```python
pow(2,500)
```

```python
pow(2,pow(2,15))
```

```python
import sys

# 增加整数字符串转换的最大位数限制
sys.set_int_max_str_digits(10000)  # 你可以根据需要调整这个值
pow(2,pow(2,15));#输出过长，省略输出结果
```

## 浮点数

python要求所有的浮点数必须带有小数部分，小数部分可以是零，这个设计可以用于区分整数和浮点数，浮点数有两种表示方法

1. 十进制表示法
2. 科学计数法
    - 96e4,4.3e-3,9.6E5
    - < a > e < b >=$a*10^b$
    
尽管浮点数0.0和整数0的含义相同，在计算机内部的表示方法也不同。

sys.float_info详细列出了python解释器运行系统的各项参数

```python
import sys
sys.float_info
```

上面输出结果中的参数：
- max：浮点所能表示的最大值
- max_exp：基数为2时最大值的幂
- max_10_exp：科学计数法最大值的幂
- dig：能准确计算的浮点数最大个数
- mant_dig：科学计数法中系数的最大精度
- epsilon：计算机所能分辨的两个相邻浮点数的最小差值

```python
sys.float_info.max
```

浮点数在超过15位的数字计算中产生的误差与计算机采用二进制有关，浮点数无法进行极高精度的运算。python支持无限制且准确的整数计算，因此对于浮点数的运算可以将小数点去掉，当做整数来计算。例如

```python
a,b=3.1415926535,1.234567898
print(a*b)
print(3.1415926535*1.234567898)
print(31415926535*1234567898)
```

#### decimal库的使用

```python
import decimal
a=decimal.Decimal('3.141592653')
b=decimal.Decimal('1.234567898')
decimal.getcontext().prec=20
a*b
```

浮点数类型的取值范围在$[2^{-1023},2^{1023}]$，运算精度为$2.220\times 10^{-16}$，对于除高精度科学计算外的绝大部分运算来说都够用
