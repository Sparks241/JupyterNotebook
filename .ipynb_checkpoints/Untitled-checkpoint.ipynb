```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.metrics import accuracy_score, classification_report
import numpy as np
file_path = '/data/workspace/data_set/套损根因数据表.csv'
data = pd.read_csv(file_path, encoding='GBK')
data
```

```python
# 2. 查看各特征的统计数据，如最大值、最小值等，查看是否存在缺失值等信息
print("数据概览：")
print(data.describe()) # 显示数值型特征的统计信息
print("\n 数据缺失值情况：")
print(data.isnull().sum()) # 显示每列缺失值的数量
```

```python
features_to_drop = ['井id','连通井id','连通水井id','连通油井id','连通井类型拼接','
套损前使用时长
','well_id_leftDf','ddsd_Sum','ddsd_Mean','ddsd_Min','ddsd_Max','ddsd_SDev','ddsd
_Median','ddsd_Count','ddsd_Variance','ddsd_FirstQuartile','ddsd_ThirdQuartile','Re
cord_Count_leftDf','well_id_rightDf','well_id']
# 使用drop 方法删除这些特征，inplace=True 表示在原地修改DataFrame
data.drop(columns=features_to_drop, inplace=True)
target = '是否套损' # 目标变量列名
features = data.columns.drop(target) # 获取所有特征列名
X = data[features]
y = data[target]
X_train, X_val, y_train, y_val = train_test_split(X, y, test_size=0.3, random_state=42)
gbdt = GradientBoostingClassifier(random_state=123456)
gbdt.fit(X_train, y_train)
gbdt.fit(X_train, y_train)
y_train_pred = gbdt.predict(X_train)
print("\n 训练集评估效果：")
print("准确率：", accuracy_score(y_train, y_train_pred))
print("分类报告：")
print(classification_report(y_train, y_train_pred))
y_val_pred = gbdt.predict(X_val)
print("\n 验证集评估效果：")
print("准确率：", accuracy_score(y_val, y_val_pred))
print("分类报告：")
print(classification_report(y_val, y_val_pred))
```

```python
from sklearn.metrics import roc_curve, roc_auc_score
# 训练集
fpr_train,tpr_train,thresholds_train = roc_curve(y_train, gbdt.predict_proba(X_train)[:,1])
roc_auc_train = roc_auc_score(y_train, gbdt.predict_proba(X_train)[:, 1])
# 验证集
fpr_val, tpr_val, thresholds_val = roc_curve(y_val, gbdt.predict_proba(X_val)[:, 1])
roc_auc_val = roc_auc_score(y_val, gbdt.predict_proba(X_val)[:, 1])
print("\n 训练集ROC AUC:", roc_auc_train)
print("验证集ROC AUC:", roc_auc_val)
import matplotlib.pyplot as plt
plt.figure()
plt.plot(fpr_train, tpr_train, color='darkorange', lw=2, label='ROC curve (train) (area= %0.2f)' % roc_auc_train)
plt.plot(fpr_val, tpr_val, color='navy', lw=2, label='ROC curve (val) (area = %0.2f)' %roc_auc_val)
plt.plot([0, 1], [0, 1], color='gray', lw=2, linestyle='--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver Operating Characteristic')
plt.legend(loc="lower right")
plt.show()
```

```python
from sklearn.metrics import precision_recall_curve
from scipy.integrate import quad
from scipy.interpolate import interp1d
import matplotlib.pyplot as plt
from sklearn.metrics import auc
precision_train, recall_train, thresholds_train = precision_recall_curve(y_train,gbdt.predict_proba(X_train)[:, 1])
precision_val, recall_val, thresholds_val = precision_recall_curve(y_val,gbdt.predict_proba(X_val)[:, 1])
plt.figure()
plt.plot(recall_train, precision_train, color='darkorange', lw=2, label='PR curve (train)(area = %0.2f)' % auc(recall_train, precision_train))
plt.plot(recall_val, precision_val, color='navy', lw=2, label='PR curve (val) (area= %0.2f)' % auc(recall_val, precision_val))
plt.xlabel('Recall')
plt.ylabel('Precision')
plt.title('Precision-Recall Curve')
plt.legend(loc="best")
plt.show()
def auc(x, y):
return quad(lambda t: interp1d(x, y, bounds_error=False, fill_value=0)(t), 0,1)[0]
```

```python
ks_stat_train = max(tpr_train - fpr_train)
ks_stat_val = max(tpr_val - fpr_val)
print("\n 训练集K-S 统计量:", ks_stat_train)
print("验证集K-S 统计量:", ks_stat_val)
```

```python
gini_train = 2 * roc_auc_train - 1
gini_val = 2 * roc_auc_val - 1
print("\n 训练集Gini 系数:", gini_train)
print("验证集Gini 系数:", gini_val)
```
