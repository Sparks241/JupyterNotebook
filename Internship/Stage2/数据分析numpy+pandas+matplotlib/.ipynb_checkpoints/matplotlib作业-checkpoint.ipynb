```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import os

current_path=os.getcwd()
data_path=current_path+"/data/taobao_data_matplolib.csv"
```

```python
df = pd.read_csv(data_path,encoding='gbk')
df.head()
```

```python
Volume=np.array(df.成交量)
a2=np.array(df.价格)
Volume
```

```python
d=np.round(list(Volume/sum(Volume)),2)
d
```

```python
grouped = df.groupby("位置")[["成交量", "价格"]].sum().reset_index()

# 计算成交量比例用于饼图
volume = grouped["成交量"]
volume_ratio = np.round(volume / volume.sum(), 2)

# 设置字体（如果有中文显示问题）
plt.rcParams['font.sans-serif'] = ['SimHei']
plt.rcParams['axes.unicode_minus'] = False
```

# 画出各省份价格、各省份成交量柱状图：

```python
plt.figure(figsize=(10, 6))
plt.bar(grouped["位置"], grouped["成交量"], color='skyblue')
plt.title("各省成交量柱状图")
plt.xlabel("省份")
plt.ylabel("成交量")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

# 画出成交量线图、柱状图、箱线图、饼图

```python
plt.figure(figsize=(10, 6))
plt.plot(grouped["位置"], grouped["成交量"], marker='o', linestyle='-', color='green')
plt.title("各省成交量折线图")
plt.xlabel("省份")
plt.ylabel("成交量")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(6, 6))
plt.boxplot(df["价格"])
plt.title("商品价格箱线图")
plt.ylabel("价格")
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8, 8))
plt.pie(volume_ratio, labels=grouped["位置"], autopct='%1.1f%%', startangle=140)
plt.title("各省成交量占比饼图")
plt.tight_layout()
plt.show()
```

# 画出价格与成交量的散点图

```python
plt.figure(figsize=(8, 6))
plt.scatter(df["价格"], df["成交量"], alpha=0.6, color='orange')
plt.title("价格与成交量散点图")
plt.xlabel("价格")
plt.ylabel("成交量")
plt.tight_layout()
plt.show()
```
