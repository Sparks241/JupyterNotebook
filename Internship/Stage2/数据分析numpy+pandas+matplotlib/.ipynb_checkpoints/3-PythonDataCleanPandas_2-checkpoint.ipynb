## 数据库风格的DataFrame连接/合并

单个键连接

```python
left = pd.DataFrame({'key': ['K0', 'K1', 'K2', 'K3'],
                      'A': ['A0', 'A1', 'A2', 'A3'],
                      'B': ['B0', 'B1', 'B2', 'B3']})
 

right = pd.DataFrame({'key': ['K0', 'K1', 'K2', 'K3'],
                       'C': ['C0', 'C1', 'C2', 'C3'],
                       'D': ['D0', 'D1', 'D2', 'D3']})
 
```

```python
left
```

```python
right
```

```python
result = pd.merge(left, right, on='key')
```

```python
pd.merge?
```

```python
result
```

多个键的连接

```python
 left = pd.DataFrame({'key1': ['K0', 'K0', 'K1', 'K2'],
                      'key2': ['K0', 'K1', 'K0', 'K1'],
                      'A': ['A0', 'A1', 'A2', 'A3'],
                      'B': ['B0', 'B1', 'B2', 'B3']})
 

 right = pd.DataFrame({'key1': ['K0', 'K1', 'K1', 'K2'],
                       'key2': ['K0', 'K0', 'K0', 'K0'],
                       'C': ['C0', 'C1', 'C2', 'C3'],
                       'D': ['D0', 'D1', 'D2', 'D3']})
 

 result = pd.merge(left, right, on=['key1', 'key2'])
```

```python
pd.merge?
```

```python
result
```

```python
result = pd.merge(left, right, how='left', on=['key1', 'key2'])
result
```

```python
result = pd.merge(left, right, how='right', on=['key1', 'key2'])
result
```

```python
result = pd.merge(left, right, how='outer', on=['key1', 'key2'])
result
```

```python
left = pd.DataFrame({'A' : [1,2], 'B' : [2, 2]})

right = pd.DataFrame({'A' : [4,5,6], 'B': [2,2,2]})

result = pd.merge(left, right, on='B', how='outer')
```

```python
result
```

合并指示符

```python
df1 = pd.DataFrame({'col1': [0, 1], 'col_left':['a', 'b']})

df2 = pd.DataFrame({'col1': [1, 2, 2],'col_right':[2, 2, 2]})

pd.merge(df1, df2, on='col1', how='outer', indicator=True)
```

也可以使用字符串作为参数替代True，字符串将会作为该列的列名称。

```python
pd.merge(df1, df2, on='col1', how='outer', indicator='indicator_column')
```

通过索引进行连接

```python
left = pd.DataFrame({'A': ['A0', 'A1', 'A2'],
                     'B': ['B0', 'B1', 'B2']},
                     index=['K0', 'K1', 'K2'])


right = pd.DataFrame({'C': ['C0', 'C2', 'C3'],
                      'D': ['D0', 'D2', 'D3']},
                      index=['K0', 'K2', 'K3'])


result = left.join(right)
left.join?
```

```python
result
```

```python
result = left.join(right, how='outer')
result
```

```python
left.join(right, how='inner')
```

```python
pd.merge(left, right, left_index=True, right_index=True, how='outer')
```

```python
pd.merge(left, right, left_index=True, right_index=True, how='inner')
```

通过索引和某列连接

```python
left = pd.DataFrame({'A': ['A0', 'A1', 'A2', 'A3'],
                     'B': ['B0', 'B1', 'B2', 'B3'],
                     'key': ['K0', 'K1', 'K0', 'K1']})


right = pd.DataFrame({'C': ['C0', 'C1'],
                      'D': ['D0', 'D1']},
                      index=['K0', 'K1'])


result = left.join(right, on='key')
```

```python
result
```

```python
pd.merge(left, right, left_on='key', right_index=True, how='left', sort=False)
```

重叠列名称的合并

```python
left = pd.DataFrame({'k': ['K0', 'K1', 'K2'], 'v': [1, 2, 3]})

right = pd.DataFrame({'k': ['K0', 'K0', 'K3'], 'v': [4, 5, 6]})

result = pd.merge(left, right, on='k')
```

```python
result
```

```python
pd.merge(left, right, on='k', suffixes=['_l', '_r'])
```

数据框拼接

```python
df1 = pd.DataFrame([[np.nan, 3., 5.], [-4.6, np.nan, np.nan],
                    [np.nan, 7., np.nan]])
 

df2 = pd.DataFrame([[-42.6, np.nan, -8.2], [-5., 1.6, 4]],
                    index=[1, 2])
```

# 数据重塑及透视表

## 多重索引

```python
columns = pd.MultiIndex.from_tuples([
        ('A', 'cat', 'long'), ('B', 'cat', 'long'),
        ('A', 'dog', 'short'), ('B', 'dog', 'short')],
    names=['exp', 'animal', 'hair_length'])


df = pd.DataFrame(np.random.randn(4, 4), columns=columns)
```

```python
df
```

```python
df.stack(level=['animal', 'hair_length'])
```

```python
df.stack(level=[1, 2])
```

```python
columns = pd.MultiIndex.from_tuples([('A', 'cat'), ('B', 'dog'),
                                     ('B', 'cat'), ('A', 'dog')],
                                    names=['exp', 'animal'])


index = pd.MultiIndex.from_product([('bar', 'baz', 'foo', 'qux'),
                                    ('one', 'two')],
                                   names=['first', 'second'])


df = pd.DataFrame(np.random.randn(8, 4), index=index, columns=columns)

df2 = df.iloc[[0, 1, 2, 4, 5, 7]]
```

```python
df
```

```python
df2
```

可以使用level参数来调用

```python
df2.stack('exp')
```

```python
df2.stack('animal')
```

```python
df3 = df.iloc[[0, 1, 4, 7], [1, 2]]
```

```python
df3
```

```python
df3.unstack()
```

```python
df3.unstack(fill_value=0)
```

```python
df3.unstack(0)
```

## 透视表

```python
import datetime

df = pd.DataFrame({'A': ['one', 'one', 'two', 'three'] * 6,
                   'B': ['A', 'B', 'C'] * 8,
                   'C': ['foo', 'foo', 'foo', 'bar', 'bar', 'bar'] * 4,
                   'D': np.random.randn(24),
                   'E': np.random.randn(24),
                   'F': [datetime.datetime(2013, i, 1) for i in range(1, 13)] +
                        [datetime.datetime(2013, i, 15) for i in range(1, 13)]})
```

```python
df
```

```python
pd.pivot_table(df, values='D', index=['A', 'B'], columns=['C'])
```

```python
pd.pivot_table(df, values='D', index=['B'], columns=['A', 'C'], aggfunc=np.sum)
```

```python
pd.pivot_table(df, values=['D','E'], index=['B'], columns=['A', 'C'], aggfunc=np.sum)
```

```python
pd.pivot_table(df, index=['A', 'B'], columns=['C'])
```

## 基本绘图：plot

Series和DataFrame上的plot方法只是围绕plt.plot（）的简单封装。

```python
ts = pd.Series(np.random.randn(1000), index=pd.date_range('1/1/2000', periods=1000))
#ts
ts = ts.cumsum()
#ts
ts.plot()
```

在DataFrame中，plot（）可以方便地绘制带有标签的所有列

```python
df = pd.DataFrame(np.random.randn(1000, 4), index=ts.index, columns=list('ABCD'))
#df
df = df.cumsum()

# plt.figure()
df.plot()
```

使用plot（）中的x和y关键字可以绘制一列与另一列的对应图形。

```python
df3 = pd.DataFrame(np.random.randn(1000, 2), columns=['B', 'C']).cumsum()

df3['A'] = pd.Series(list(range(len(df))))
#df3
df3.plot(x='A', y='B')
```

## 其他绘图

绘制方法允许除了默认线条图之外的其他绘图样式。这些方法可以通过plot（）的kind关键字参数控制。包括以下内容：  

- bar 或 barh 绘制条形图
- hist 绘制 直方图
- box 绘制 箱线图
- kde 或 density 绘制 密度图
- area 绘制 区域图
- scatter 绘制 散点图
- hexbin 绘制 hexagonal bin plots
- pie 绘制 饼图

除了使用kind参数，你也可以使用直接对应的方法来完成相应的绘图效果。

### 条形图

```python
df
```

```python
df.iloc[5].plot(kind='bar')
plt.axhline(0, color='k')
```

调用DataFrame的plot.bar（）方法生成一个多个条形图

```python
df2 = pd.DataFrame(np.random.rand(10, 4), columns=['a', 'b', 'c', 'd'])
print (df2)
df2.plot.bar()
```

要生成堆叠的条形图， 传入：stacked=True:

```python
df2.plot.bar(stacked=True)
```

使用barh方法获得水平条形图：

```python
df2.plot.barh(stacked=True)
```

### 直方图

直方图可以通过 DataFrame.plot.hist() 和 Series.plot.hist() 进行绘制。

```python
df4 = pd.DataFrame({'a': np.random.randn(1000) + 1, 'b': np.random.randn(1000),
                    'c': np.random.randn(1000) - 1}, columns=['a', 'b', 'c'])

#df4.head()
# plt.figure()

df4.plot.hist(alpha=0.5)
```

```python
?df4.plot.hist
```

直方图可以通过参数实现堆叠 stacked=True。 宽度大小可以通过bins关键字更改。

```python
df4.plot.hist(stacked=True, bins=20)
```

### 箱线图

```python
df = pd.DataFrame(np.random.rand(10, 5), columns=['A', 'B', 'C', 'D', 'E'])
df
df.plot.box()
```

通过传递颜色关键字可以对Boxplot进行着色。

```python
color = dict(boxes='DarkGreen', whiskers='DarkOrange',
             medians='DarkBlue', caps='Gray')


df.plot.box(color=color, sym='r+')
```

通过vert来控制横向，以及通过positions控制每个图的位置

```python
df.plot.box(vert=False, positions=[1, 4, 5, 6, 8])
```

也可以通过boxplot的方法直接绘图

```python
df = pd.DataFrame(np.random.rand(10,5))
df
df.boxplot()
```

您可以使用by关键字参数创建分层的boxplot来创建分组绘制图形

```python
df = pd.DataFrame(np.random.rand(10,2), columns=['Col1', 'Col2'] )

df['X'] = pd.Series(['A','A','A','A','A','B','B','B','B','B'])
df
df.boxplot(by='X')
```

```python
df = pd.DataFrame(np.random.rand(10,3), columns=['Col1', 'Col2', 'Col3'])

df['X'] = pd.Series(['A','A','A','A','A','B','B','B','B','B'])

df['Y'] = pd.Series(['A','B','A','B','A','B','A','B','A','B'])
print (df)
df.boxplot(column=['Col1','Col2'], by=['X','Y'])
```

### 散点图

```python
df = pd.DataFrame(np.random.rand(10, 4), columns=['a', 'b', 'c', 'd'])

df.plot.scatter(x='a', y='b')
```


如果希望再一个图上绘制多个列，指定绘图的目标ax后可以重复调用绘图方法，建议指定颜色和标签关键字来区分每个组。

```python
print (df)
df["e"]=["k","k","k","k","k","k","k","k","k","k"]
ax = df.plot.scatter(x='a', y='b', color='DarkBlue', label='Group 1')

df.plot.scatter(x='c', y='d', color='r', label='Group 2', ax=ax)
```

```python
?df.plot.scatter
```

关键字c可以作为列的名称给出，以为每个点提供颜色

```python
df.plot.scatter(x='a', y='b', c='c', sharex=False, s=50)
```

```python
df.plot.scatter(x='a', y='b', s=df['c']*200)
```

### Scales

可以通过传入 logy 获得 log-尺度的 Y 轴。

```python
ts = pd.Series(np.random.randn(1000), index=pd.date_range('1/1/2000', periods=1000))

ts = np.exp(ts.cumsum())

ts.plot(logy=True)
```

### 在双Y轴上面画图

```python
df.A.plot()

df.B.plot(secondary_y=True, style='g')
```

要在DataFrame中绘制一些列，请将列名称给secondary_y关键字

```python
ax = df.plot(secondary_y=['A', 'B'])

ax.set_ylabel('CD scale')

ax.right_ax.set_ylabel('AB scale')
```

注意，在辅助y轴上绘制的列在图例中自动标记为“（右）”。要关闭自动标记，请使用mark_right = False关键字：
